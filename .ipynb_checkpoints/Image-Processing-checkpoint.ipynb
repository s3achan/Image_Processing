```python
from numpy  import *
from PIL import Image,ImageFilter
import matplotlib.pyplot as plt
```

```python
photo = 'redbull.jpg'
out = 'saved.jpg'
```

```python
img=Image.open(photo)
plt.imshow(img)
plt.show()
img.size
```

```python
img=Image.open(photo)
img.save(out)
img=Image.open(out)
plt.imshow(img)
plt.show()
```

```python
region = (100,100,700,450)
cropImg = img.crop(region)
plt.imshow(cropImg)
plt.show()
```

```python
img_rotate = img.rotate(45)
plt.imshow(img_rotate)
plt.show()
```

```python
img_rotate = img.transpose(Image.ROTATE_90)
plt.imshow(img_rotate)
plt.show()
```

```python
img_resize = img.resize((400, 300))
plt.imshow(img_resize)
plt.show()
```

```python
img_convert = img.convert('L')
plt.imshow(img_convert)
plt.show()
```

```python
WHITE, BLACK = 255, 0 
img_point = img.point(lambda x: WHITE if x > 128 else BLACK) 
img_binary = img_point.convert('1')  
plt.imshow(img_binary)
plt.show()
```

```python
img_filter = img.filter(ImageFilter.EMBOSS)
plt.imshow(img_filter)
plt.show()
```

```python
img_filter = img.filter(ImageFilter.EDGE_ENHANCE)
plt.imshow(img_filter)
plt.show()
```

```python
img_filter = img.filter(ImageFilter.CONTOUR)
plt.imshow(img_filter)
plt.show()
```
